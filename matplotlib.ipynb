```python
import numpy as np
import matplotlib.pyplot as plt
x =np.arange(0,10)
y = 2* x + 5
z =np.sin(x)
#changing the figure
fig =plt.figure(figsize = (5,5))
#plotting the data points
plt.plot(x,y,linewidth = 1.0,linestyle = ':',color ='g',alpha =1,marker ='o',markerfacecolor = 'blue',markersize = 8,markeredgewidth = 2)
plt.plot(x,z,linewidth = 1.0,linestyle = (0, (3, 10, 1, 15)),color ='r',alpha =0.8,marker =6,markerfacecolor = 'black',markersize = 8,markeredgewidth = 2)
plt.plot(y,z,linewidth = 1.0,linestyle = (5, (10, 3)),color ='r',alpha =0.8,marker =3,markerfacecolor = 'blue',markersize = 8,markeredgewidth = 2)
plt.title('line graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(['xylable','xzlable','yzlable'],loc = 'best')
plt.grid(True)
plt.show()
```

**sub plots**

```python
x =np.arange(0,10)
y1 = 2* x + 5
y2 =np.sin(x)
y3 = 3 *x +10
y4 = 0.5 *x**2
#creating a 2x2 matrix sub plots
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('1st graph')
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('2nd graph')
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('3rd graph')
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('4th graph')
#tight_layout gives better output(no over lapping)
plt.tight_layout()
plt.show()
```

```python
import mplfinance as mpf
import pandas as pd
data = {
    'data': ['2024-08-31','2024-08-30','2024-08-29'],
    'open': [100,105,103],
    'high': [110,108,108],
    'low' : [95,100,101],
    'close' : [105,103,107],
    'volume': [1000,1200,1100]
}
# convert data to a dataframe
df = pd.DataFrame(data)
df['data'] = pd.to_datetime(df['data'])
df.set_index('data',inplace= True)
fig = plt.figure(figsize = (0,0))
#create a candle stick chart with volume
mpf.plot(df,type ='candle',style = 'charles',title = 'candlestick chart',ylabel = 'price',volume =True)
```
