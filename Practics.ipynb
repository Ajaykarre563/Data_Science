**creating dataframe from a sql query**

```python
import sqlite3
conn = sqlite3.connect(':memory:')
conn.execute('''
CREATE TABLE people(
  id integer primary key,
  name text,
  age integer
)''')
conn.execute('''
INSERT INTO people(name,age)
VALUES ('alice',20),
('bob',30)
''')
query = 'select * from people'
df  = pd.read_sql_query(query,conn)
df
```

```python
import sqlite3
import pandas as pd
conn =sqlite3.connect(":memory:")
conn.execute('''
CREATE TABLE people3(
  id integer primary key,
  name text,
  age integer
)''')
conn.execute('''
INSERT INTO people3(name, age)
VALUES('san',21),
('shiv',27)''')
query = 'select * from people3'
df =pd.read_sql_query(query,conn)
df.index.name = 'sn'
df.index = ["->","->"]
df
```

**imputings**

```python
import pandas as pd
data = {
    'A': [1, 2, None, 4],
    'B': [None, 5, None, 7],
    'C': [9, None, 11, None]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
new_data =df.copy()
new_data
```

```python
from sklearn.impute import SimpleImputer
mean_imputer =SimpleImputer(strategy='mean')
new_data.iloc[:,:] = mean_imputer.fit_transform(new_data)
```

```python
new_data
```

```python
new_data.isnull().sum()
```

**imputing with maual given dataset**

```python
#filling a constant value of a paticular column
new_data['A'].fillna('A',inplace=True)
```

```python
new_data
```

```python
#filling a constant value of a paticular column
new_data['B'].fillna(4,inplace=True)
```

```python
new_data
```

```python
#filling with a mean value of a particular colummn
new_data['C'].fillna(new_data['C'].mode(),inplace =True)
```

```python
new_data
```

**forward filling**

```python
new_data.fillna(method ='ffill',inplace =True)
new_data['A'][0:4]
```

```python
new_data
```

**backward filling**

```python
new_data.fillna(method ='bfill',inplace =True)
new_data['B'][0:4]
```

**interpolation**

```python
new_data.interpolate(limit_direction="both",inplace =True)
new_data['C'][0:4]
```
